```python
import time
import tracemalloc
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.svm import  SVC
import numpy as np
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv("dataset.csv")
df.info()
df
```

```python
X=df.iloc[:,1:5]
y=df['Species']
```

```python
X.tail()
```

```python
y.tail()
```

```python
svm = SVC(kernel='linear', random_state=1, gamma=0.001, C=10)

svm.fit(X, y)
```

**Accuracy Score**

```python
y_pred=svm.predict(X)
print(accuracy_score(y, y_pred))
```

**Runtime Fit**

```python
runtimefit = []

for _ in range(100):
    start = time.perf_counter()
    svm.fit(X, y)
    stop = time.perf_counter()
    runtimefit.append((stop - start) * 1000)
```

```python
runtimefit_df = pd.DataFrame(runtimefit, columns=["Milliseconds"])
```

```python
runtimefit_df.describe()
```

**Runtime Predict**

```python
runtimepredict = []

for _ in range(100):
    start = time.perf_counter()
    y_pred=svm.predict(X)
    stop = time.perf_counter()
    runtimepredict.append((stop - start) * 1000)
```

```python
runtimepredict_df = pd.DataFrame(runtimepredict, columns=["Milliseconds"])
```

```python
runtimepredict_df.describe()
```

**Memory Consumption**

```python
tracemalloc.start()
svm.fit(X, y)
print(tracemalloc.get_traced_memory())


tracemalloc.stop()
```
